```python
#AtividadeContextualizadaAula6
#Programa para procedimento cirúrgico

print('Olá! Bem vindo ao programa para procedimento cirúrgico da aula 6.')

#Etapa 1
  #Etapa 1.1 - Procedimento de anestesia: tipo de anestesia a ser utilizada e dosagem em ml.
while True:
  #Será solicitada a anestesia a ser utilizada.
  #Se o pesquisador escolher uma anestesia inválida, o programa repete a pergunta e não passa para a próxima etapa.
  anestesia = str(input('A anestesia utilizada será ketamina e xilazina, em conjunto, ou halotano?'))
  if (anestesia=='ketamina e xilazina'):
    anestesia = int(input('Digite a dosagem desejada para o animal, em ml, levando em consideração seu peso:'))
    break
  if (anestesia=='halotano'):
    anestesia = int(input('Digite a dosagem desejada para o animal, em ml, levando em consideração seu peso:'))
    break
  else:
    print('Escolha uma anestesia correta para o animal.')

#próxima etapa


  #Etapa 1.2 - Anestesia fez efeito no animal
    #O animal deve estar anestesiado, bem posicionado, e com as barras que suportam seu peso posicionadas no ouvido externo do animal.
while True:
  posic= str(input('A anestesia já fez efeito?\n'))
  if(posic=='nao'):
    print('Aguarde até que o animal esteja anestesiado para iniciar a cirurgia.')
  else:
    print('Ok, então, vamos dar continuidade ao procedimento.')
    break

#próxima etapa


#Etapa 2 - Verificação de posicionamento do animal
while True:
  barra = str(input('As barras que suportam o peso do animal estão posicionadas no ouvido externo do animal?'))
  if (barra=='nao'):
    print('Posicione-as corretamente no ouvido externo do animal para dar continuidade ao procedimento.')
  else:
    barra2 = str(input('O animal piscou após o posicionamento da barra?')) 
    #Essa piscadela acontece devido ao estímulo da musculatura responsável por este movimento.
    if (barra2=='sim'):
      print('Então, verifique se a angulação da cabeça do animal está sem diferenças entre bregma e lambda.')
      print('Isto é necessário para garantir uma superfície de cirurgia plana.')
      break
    else:
      print('Ele deve piscar. Verifique se houve realmente o posicionamento correto da barra.')

#próxima etapa


#Etapa 3 - Limpeza do campo de trabalho
  #Etapa 3.1 - Retirada da pelagem que recobre a parte superior da calota craniana
  #Etapa 3.2 - Retirada dos tecidos moles (epiderme, derme e tecido conjuntivo) até alcançar a parte óssea da caixa craniana
  #Etapa 3.3 - Limpeza da calota craniana utilizando H2O2 10 volumes.

print('Podemos iniciar o procedimento. Inicie retirando a pelagem que recobre a parte superior da calota craniana.')
  #Aqui será solicitado que o pesquisador retire toda a pelagem.
  #Se ela não for retirada, a pergunta se repete até que o pesquisador indique que retirou.
while True:
  limp1 = str(input('Você retirou toda a pelagem do local indicado?')) #CHECAR ESSA CÉLULA AMANHÃ!!!!!!! ela fica repetindo infinitamente
  if (limp1=='nao'):
    print('Realize a retirada da pelagem para dar continuidade ao procedimento.')
  elif (limp1=='sim'):
    print('Então, realize a retirada dos tecidos moles, até que alcance a parte óssea da caixa craniana.')
    break

while True:
  #Aqui será solicitado que o pesquisador retire os tecidos moles.
  #Se eles não forem retirados, a pergunta se repete até que o pesquisador indique que retirou.
  limp2 = str(input('Toda a epiderme, derme e tecido conjuntivo foram retirados?')) 
  if (limp2=='nao'):
    print('Realize a retirada de todos os tecidos moles para dar continuidade ao procedimento.')
          #O pesquisador precisa tirar tudo para seguir para a última parte da limpeza.
  elif (limp2=='sim'):
    print('Então, realize a limpeza da calota craniana utilizando H2O2 10 volumes.')
    break

while True:
  #Aqui será solicitado que o pesquisador limpe totalmente a calota craniana.
  #Enquanto ela não estiver limpa, a pergunta se repete.
  limp3 = str(input('A calota craniana está limpa, sem resquícios de tecidos moles?'))
  if (limp3=='nao'):
    print('Realize a limpeza novamente.')
  if (limp3=='sim'):
    print('Então, vamos dar continuidade ao procedimento.')
    break

#próxima etapa


#Etapa 4 - Preparação do animal para evitar sangramentos 
while True:
  #Será solicitado que o pesquisador coloque poliacrilato no local do procedimento para evitar sangramentos.
  poli = str(input('O animal está bem posicionado e o campo de cirurgia está limpo?'))
  if (poli=='sim'):
    print('Então, utilize uma pequena camada de poliacrilato em todo o perímetro externo para evitar sangramentos.')
    break
  if (poli=='nao'):
    print('O animal precisa estar bem posicionado e com o campo de cirurgia limpo. Repita os processos.')

#próxima etapa


#Etapa 5 - Escolha de ponto para a fixação de parafusos
print('Depois, escolha um ponto para fixação dos parafusos.')
print('A camada óssea é mais espessa e suporta uma maior profundidade do parafuso.')
print('(Parafusos maiores devem dar até 3 voltas.)')

AP=6.4 #AnteroPosterior
LL=3.2 #LateroLateral
DV=4.2 #DorsoVentral

while True:
  posicIndicado = str(input('A agulha está posicionada sobre o bregma?'))
  if (posicIndicado=='sim'):

    dicionarioAP={'AP': [6.42-0.42]}
    dicionarioLL1={'LLdir': [3.33+0.30]}
    dicionarioLL2={'LLesq': [3.33-0.30]}
    dicionarioDV={'DV': [4.20-0.20]}
      #Os dicionários foram feitos para os cálculos de cada ponto.

    print('Então: o valor utilizado para o ponto AnteroPosterior é:', dicionarioAP)
    print('O valor utilizado para o ponto LateroLateral direito é:', dicionarioLL1)
    print('O valor utilizado para o ponto LateroLateral esquerdo é:', dicionarioLL2)
    print('O valor utilizado para o ponto DorsoVentral:', dicionarioDV)
    print('Você utilizará esses valores posteriormente. Guarde-os.')
          #Próxima etapa
    break
    #Se a agulha estiver corretamente posicionada, o programa já calculará para o usuário os valores para uso no hipocampo em cada ponto.
    #Os cálculos de hipocampo usados foram os valores de cada ponto subtraídos ou somados à outros já designados na atividade:
      #AP: 6,42 cm – 0,42 = 6,00
      #LLdir: 3,33 cm + 0,30 = 3,63
      #LLesq: 3,33 cm - 0,30 = 3.03
      #DV: 4,20 cm – 0,20 = 4,00
  else:
    print('Posicione a agulha sobre o bregma para prosseguir.')

#próxima etapa


#Etapa 6 - Localizar os pontos de inserção das cânulas-guia
  #A localização destes pontos é definida pelos valores encontrados nos cálculos da etapa anterior.
while True:
  valueAP = float(input('Indique a posição do ponto AnteroPosterior:'))
        #O pesquisador utilizará os valores exatos que foram encontrados na etapa anterior.
            #AP: 6,42 cm – 0,42 = 6,00
            #LLdir: 3,33 cm + 0,30 = 3,63
            #LLesq: 3,33 cm - 0,30 = 3.03
  if valueAP==6.0:
    print('Agulha corretamente posicionada no ponto AnteroPosterior.')
    break
  else:
    print('Insira o valor correto para que ocorra o posicionamento.')

while True:
  valueCanula = str(input('A primeira cânula será colocada no hemisfério direito ou esquerdo?'))
    #A depender da escolha do hemisfério, o pesquisador deverá indicar os valores para o LL esquerdo e direito.
    #Foram colocados valores com até dois números após a vírgula.
  if valueCanula == 'direito':
    valueLL1 = float(input('Indique a posição do ponto LateroLateral direito:'))
    if valueLL1==3.63:
      print('Agulha corretamente posicionada no ponto LateroLateral direito.')  
    else:
      print('Insira o valor correto para que ocorra o posicionamento.')

    valueLL2 = float(input('Indique a posição do ponto LateroLateral esquerdo:'))
    if valueLL2==3.03:
      print('Agulha corretamente posicionada no ponto LateroLateral esquerdo.')
      break
    else:
      print('Insira o valor correto.')

  if valueCanula == 'esquerdo':
    valueLL2 = float(input('Indique a posição do ponto LateroLateral esquerdo:'))
    if valueLL2==3.03:
      print('Agulha corretamente posicionada no ponto LateroLateral esquerdo.')
    else:
      print('Insira o valor correto.')

    valueLL1 = float(input('Indique a posição do ponto LateroLateral direito:'))
    if valueLL1==3.63:
      print('Agulha corretamente posicionada no ponto LateroLateral direito.')
      break 
    else:
      print('Insira o valor correto para que ocorra o posicionamento.')

#próxima etapa


#Etapa 7 - Alcançar as meninges (com broca)
print('Agora, você deve furar o crânio com a broca até alcançar as meninges.')
while(True):
  perfMeninges = int(input('Indique o ângulo, em graus, para perfuração:'))
    #O ângulo ideal de perfuração é 45.
    #Se o pesquisador utilizador outro ângulo, o programa avisará que o ângulo deve ser reajustado.
  if perfMeninges==45:
    print('Angulação correta. Realize o furo, com cuidado, até alcançar as meninges.')
    print('Para evitar perfurá-las, apoie a mão que segura a broca contra o estereotáxico.')
    break
  else:
    print('Indique o ângulo correto para uma boa perfuração.')
    continue

#próxima etapa


#Etapa 8 - Introdução da cânula-guia ao ponto DV
print('Na próxima etapa, introduza a cânula-guia no ponto DorsoVentral anteriormente calculado.')
while(True):
  valueCGuiaDV = float(input('Indique a posição do ponto DorsoVentral:')) #DV = 4.00
  if valueCGuiaDV==4.00:
    print('Perfeito. Introduza a cânula-guia.') 
    break
  else:
    print('Indique o valor correto para introdução da cânula-guia.')
    continue 

#próxima etapa


#Etapa 9 - Drenar qualquer sangue ou líquido cefalorraquidiano que esteja saindo
print('Feita a introdução, drene qualquer sangue ou líquido cefalorraquidiano, caso esteja saindo.')
print('Para isso utilize pequenos rolos de papel absorvente.')

listaLimpLiq = ['Líquidos saindo:','Sangue', 'Líquido Cefalorraquidiano']
for i in listaLimpLiq:
  print(i)


#próxima etapa


#Etapa 10 - Mistura do acrílico polimerizante com o solvente (capacete)
print('Agora, você deve fazer uma mistura do acrílico polimerizante com o solvente.')
print('Esta mistura deve possuir textura espessa porém maleável.')

while True:
  cap1 = input('A mistura feita está com a textura ideal?')
  if cap1=='sim':
    while(True):
      cap2 = input('Ela é suficiente para cobrir crânio, cânula-guia e parafuso?')
      if cap2=='sim':
        print('Espere até que ela esteja seca e rígida.')
        break
      else:
        print('Faça mais dessa mistura, ela deve ser suficiente para cobrir a área desejada.')
        continue
    break
  else:
    print('Atinja a textura ideal antes de continuar.')
    continue  


#próxima etapa


#Etapa 11 - Fixação da outra cânula-guia
print('Com a mistura seca, levante levemente o braço do estereotáxico para fixar a segunda cânula-guia.')
print('Tenha cuidado para não movimentar a primeira que foi fixada.')

while(True):
  valueCGuiaDV2 = float(input('Indique a posição do ponto DorsoVentral:')) #DV = 4.00
  if valueCGuiaDV2==4.00:
    print('Introduza a segunda cânula-guia.') 
    break
  else:
    print('Indique o valor correto para introdução da segunda cânula-guia.')
    continue


#próxima etapa


#Etapa 12
  #Etapa 12.1 - Drenar qualquer sangue ou líquido cefalorraquidiano que esteja saindo
print('Drene novamente qualquer sangue ou líquido cefalorraquidiano, caso esteja saindo.')

while(True):
  valueLimpCGuia2 = input('Há sangue ou líquido cefalorraquidiano saindo?') 
  if (valueLimpCGuia2=='sim'):
    print('Então, limpe o local com rolos de papel absorvente.') 
    continue
  else:
    print('Então, faça novamente, a mistura do acrílico polimerizante com o solvente.')
    break

  #Etapa 12.2 - Mistura do acrílico polimerizante com o solvente (capacete 2)
while True:
  cap3 = input('A mistura feita está com a textura ideal?') #Se estiver com a textura certa, ele segue para a pergunta de quantidade
  if cap3=='sim':
    while(True):
      cap4 = input('Ela é suficiente para cobrir crânio, cânula-guia e parafuso?') #Se for suficiente, passa a etapa
      if cap4=='sim':
        print('Espere até que ela esteja seca e rígida.')
        print('Deixe sempre  um espaço entre o capacete e o início da área tecidual.')
        break
      else: #Se não for suficiente, solicita que faça mais mistura
        print('Faça mais dessa mistura, ela deve ser suficiente para cobrir a área desejada.')
        continue
    break
  else: 
    #Se a mistura não tiver a textura certa, ele solicita que atinja a textura ideal.
    print('Atinja a textura ideal antes de continuar.')
    continue  


#próxima etapa


#Etapa 13 - Acomodação do animal
print('Com a mistura seca, levante novamente, com cuidado o braço do estereotáxico.')
print('Não movimente as cânulas-guias fixadas.')

while(True):
  valueFimCirurgia = input('Você finalizou a cirurgia?') 
  if valueFimCirurgia=='sim':
    print('Então vamos finalizar o procedimento.') 
    print('Acomode o animal em uma caixa aquecida por uma lâmpada e sem outros animais acordados.')
    break
  else:
    print('Finalize o que estiver faltando antes de seguir para a acomodação do animal.')
    continue

while(True):
  valueSleep = input('O animal acordou?') #o animal só deve voltar à sua caixa-moradia quando acordar.
  if valueSleep=='sim':
    print('Então ele está pronto para retornar à sua caixa-moradia.') 
    break
  else:
    print('Aguarde até que ele acorde para poder retorná-lo à sua caixa moradia.')
    continue

print('Fim da cirurgia')

#finalização
#dias de glória
```
